# Report: How people's talk about novel technological innovations?


## Introduction

Kuznets & Murphy (1966) argued more than 30 years ago that modern economic growth is intricately tied to the expansion of technology, as highlighted in his work with Murphy (1966). Recognizing that technological innovations play a pivotal role in the generation and dissemination of knowledge; it becomes imperative to delve into an examination of people's attitudes towards these innovations throughout history. By understanding how individuals have perceived and embraced technological advancements, we gain valuable insights into the dynamics that drive economic growth, laying the foundation for informed analyses of historical developments and shaping future trajectories in technology and society.

This project serves as a pilot study to investigate how people talk about novel technological innovations utilizing a diachronic corpus spanning from 1710 to 1920 to study the people’s attitudes on some of the major technological innovations during the first and second industrial revolutions. The two research questions proposed here are:

1. Do people tend to talk more or less about novel innovations as time goes by?
2. Do people tend to have positive or negative attitudes towards novel innovations as time goes by?

The hypothetical answer to research question 1) is that: yes, people tend to talk more about one novel innovation for a certain period after the inventions and the frequency will decrease after that as people get more used to it. Furthermore, based on the results from (Michel et al., 2011) which suggest that the pace of cultural assimilation of technology has accelerated, we hypothesise that this process will become faster as time goes by.

The hypothetical answer to research question 2) is that, at the beginning, people will have negative attitudes towards one innovation and it will change towards positive after the adoption and widely utilizations. This hypothesis is based on (Mokyr & Strotz, 1998) that the adaptation of a technology can be inflexible and slow to improve and people tend to resist changes brought by new technology. 

## Methodology
To test the hypothesis proposed before, frequency analysis and sentiment analysis are done for The Corpus of Late Modern English Texts, version 3.1 (CLMET 3.1) (Diller et al., 2011). CLMET3.1 represents a systematic compilation of public domain texts sourced from diverse online archiving initiatives. The corpus encompasses approximately 34 million words of continuous text and exclusively features compositions authored by native English speakers from Britain and Ireland. Notably, the corpus adheres to the principle of limiting the inclusion of texts by any single author to a maximum of three. This corpus is genre-balanced with temporal metadata annotated to the granularity of ten years. The detail of word count division is shown in Table 1 (Diller et al., 2011).


![Table 1](Table_1.jpg) 

The pre-processing of the text includes the following. All texts are transformed into lower cases to retrieve all sentences containing the inventions of interest. Unnecessary tags such as \<p> are removed. Sentence segmentation is done through the Spacy module (Honnibal & Montani, 2017). After pre-processing, sentences containing the technological innovations are extracted by identifying words of interest. 

For the first industrialization, the innovations chosen are as follows. 
a.	The invention of steam engine and steam power: its significance lies in the fact that the transformation of stationary steam engines into mechanisms capable of rotary motion rendered them apt for industrial applications, which was seen as the ignition for industrialization (Landes, 2003); 
b.	the invention of machine tools in general: Machine tools facilitated the cost-effective production of precise metal components, albeit requiring several decades to devise efficient methods for manufacturing interchangeable parts (Allen, 1978);
c.	The Spinning Jenny: according to (Broadberry & Gupta, 2005), industrialization starts with mechanized textiles spinning in the 1780s.

For the second industrialization, the innovations chosen are as follows.
a.	Iron and steel: By 1850, the iron era was established. In 1856, Henry Bessemer made it possible to make cheap steel, which boosted the growth of the steel industry and has come into the popular mind to symbolize the technology of the Second Industrial Revolution (Mokyr & Strotz, 1998).
b.	Railways: In the Second Industrial Revolution, railroads advanced in speed, safety, and comfort through incremental innovations (Mokyr & Strotz, 1998).
c.	Bicycle: it is novel and special in the sense that it didn't substitute an existing method but emerged as a mass transportation tool, profoundly influencing urban residential patterns with immeasurable consequences.
d.	Telegraph: it is the first effective application of electricity which is marked as a technological triumph that has reshaped communication (Mokyr & Strotz, 1998).

For frequency analysis, the normalized frequency of the total number of sentences containing the words of interest in its total number of sentences in each decade is calculated to minimize the effect that the number of texts in each decade is not balanced. The normalized frequency is sorted by chronological order and the line chart is generalized to observe its changing pattern.

For sentiment analysis, the sentiment score of each sentence is calculated using Google Cloud Natural Language API. This is a pre-trained model that had been utilized in diachronic corpus analysis such as (Mayer et al., 2022), therefore, it is chosen for its better suitability for data in this project compared to other rule-based models trained on modern web data such as Vader.  The mean, median, max and min sentiment scores are calculated for sentences in each time bin. There is no drastic variation in the pattern of the four types of statistic numbers. Visualizations of mean and median are shown in the form of line charts in yearly sorted order. 

## Findings and Discussion

### Frequency analysis

Overall, the changing pattern of frequency shown in each word of interest supports the hypothesis that the frequency increases after the invention and decreases after it is well adopted in society. For the invention of the steam engine/steam power, where there is only a slight increase in frequency after the 1750s and a drastic increase is shown after the 1820s.

For inventions in the second industrialization, such as the telegraph, railway, bicycle, and electricity. There was an apparent increase right after the invention and the frequency fluctuated more than steam engine/steam power that it goes down immediately after the invention. Since the steam engine/steam power was invented earlier, it supports the hypothesis that the process of increase first and then decrease is faster, which conforms with the claim that the assimilation of technology accelerated.

This project proposes several potential reasons to account for this phenomenon. The first reason is that steam engine/steam power is less close to an individual’s daily life. Since this corpus is genre-balanced it not only includes technological-related text, there can be less discussion of breakthroughs that have not yet affected common people’s life in an evident way. The second reason is that the adoption of steam engine/steam power in industry has been relatively slow compared to inventions in the second industrialization, which potentially accounts for the phenomenon that the discussion has become more frequent after its wide adoption. The third reason to account for this is that the application of modern communication methods accelerated the spread of information and further enabled the instant reaction to novel inventions reflected through discussion in texts. The above reasons can serve as potential hypotheses to be further tested through other analysis methods such as content analysis through semantic tagging.

### Sentiment analysis

Overall, the line charts of steel, bicycle, and telegraph support the hypothesis that people have negative attitudes towards novel technology innovations and this change to positive later. This suggests the potential fact that people only accept a new technology after having a negative attitude towards it. it is noteworthy that the negative scores are more extreme than the positive scores. While the negative score can go down to -0.8, the positive score remains around 0.2-0.4. The potential reason behind this is that, after the wide adoption of certain technology, people are getting more used to its presentation in daily life without having strong emotions about it.

## Conclusion and Limitation

In conclusion, this pilot study utilizing a diachronic corpus spanning from 1710 to 1920 sheds light on the changing patterns in people's attitudes towards technological innovations during the first and second industrial revolutions. The frequency analysis reveals a consistent trend supporting the hypothesis that discussions about novel innovations increase after their invention and subsequently decrease as they become widely adopted. The faster assimilation observed in later inventions aligns with the notion that technological assimilation has accelerated over time. Moreover, sentiment analysis provides valuable insights into the emotional trajectory associated with technological advancements. The general pattern indicates initial negative attitudes shifting towards positivity after widespread adoption. 

However, it is crucial to acknowledge the limitations of this study. The corpus, while comprehensive, may not capture the entirety of public sentiment due to its genre balance, potentially underrepresenting discussions not directly related to technological breakthroughs. Additionally, sentiment analysis tools inherently simplify complex emotions, and nuances in language and context may not be fully captured. The study's focus on English texts from Britain and Ireland limits its generalizability to a broader global context. Further research, incorporating diverse languages and cultures, is essential for a more comprehensive understanding of global attitudes towards technological innovations.

```python
#A function to loop over a folder and retrieve all text and store them as list of strings
def get_texts(folder_path):
    '''
    input: folder_path
    process: loop through the text files in the folder; read all text in each text file as one string; store them in one list
    output: return a list of strings, each string is one text file
    '''
    texts = []
    for file_name in os.listdir(folder_path):
        if file_name.endswith('.txt'):  # Check if the file is a text file
            file_path = os.path.join(folder_path, file_name)  # Get the full file path
            with open(file_path, 'r') as file:
                # Perform operations with the file content here
                content = file.read()
                texts.append(content)
    return texts
```

```python
def remove_punctuation(text):
    # Create a translation table mapping all punctuation characters to None
    translator = str.maketrans('', '', string.punctuation)
    
    # Use translate() to remove punctuation characters
    text_without_punctuation = text.translate(translator)
    
    return text_without_punctuation
```

```python
#A function to pre-process a single text
def pre_process(text):
    '''
    input: text(a large string) For this research, one single document
    get rid of unwanted tags in the text
    transform all words into lower case spelling
    outpyt: cleaned_up_text
    '''
    
    clean_text = text.lower().replace("<p>", " ").replace("</p>", " ").strip("\n")
    return clean_text
```

```python
#A function to extact content between the tags
def extract_string_between_tags(text, TOI):
    '''
    input: text(string), TOI(string:tag of interest)
    process: retrieve content within the matched tag using regular expressions
    output: the string between the tag
    '''
    pattern = r'<{}>(.*?)<\/{}>'.format(TOI, TOI)  # Regular expression pattern to match content between <decades> tags
    matches = re.search(pattern, text)  # Searching for the pattern in the input string

    if matches:
        extracted_string = matches.group(1)  # Extracting the content captured within the first group of parentheses
        return extracted_string
    else:
        return None  # Return None if no match is found
```

```python
#A function to pick out word of interest wanted in the text and put the sentence into time-bins
def sent_time(text_of_sent, WOI):
    '''
    input: a folder of clean texts (a list of lists, each lists consists of strings), WOI (a string: a word of interest)
    process: 1. sentence segamentation 
             2. find sentence that contain the word of interest 
             3. put the sentence into its time bins
    return: a dictionary (WOI should be specified in the variable name) with keys being its decades and sentences being its elements.
    '''
    time_bin = {}
    for text in text_of_sent:
        #print(text[0])
        #print(extract_string_between_tags(text[0], "id"))
        for sent in text:
            #sent_stem = stem_words_spacy(sent)
            if WOI in sent:
                decade = extract_string_between_tags(text[0], "decade")
                if decade in time_bin:
                    time_bin[decade].append(sent)
                else:
                    time_bin[decade] = []
                    time_bin[decade].append(sent)
        
    return time_bin
    
```

```python
#calculate total number of sentences in each decades
def calc_total(text_of_sent):
    '''
    input: text_of_sent(list of lists)
    process: extract inner list's first element which contain decade tag, count the total number of sentences
             add to the total number in each decade
    output: a dictionary for the total number of sentences(values) in each decade(keys)
    '''
    sub_total = {}
    for l in text_of_sent:
        decade = extract_string_between_tags(l[0], "decade")
        if decade in sub_total:
            sub_total[decade] += len(l)
        else:
            sub_total[decade] = len(l)
    return sub_total
```

```python
def analyze_sentiment(text_content: str = "I am so happy and joyful.") -> float:
    """
    Analyzes Sentiment in a string.

    Args:
      text_content: The text content to analyze.
      
    Returns:
      Sentiment score of the text content.
    """

    client = language_v2.LanguageServiceClient()

    document = language_v2.Document(content=text_content, type_=language_v2.Document.Type.PLAIN_TEXT,  language_code = "en")
    response = client.analyze_sentiment(document=document)

    return response.document_sentiment.score

# Example usage:
sentiment_score = analyze_sentiment("I am feeling bad today!")
print(f"Sentiment score: {sentiment_score}")
```

```python
def plot_multi_lines(mean_d, median_d, max_d, min_d, word):
    '''
    input: four dictionaries of different statistics scores, word: strings (WOI), #tp: string ("sent")
    process: set the line chart values, size, grid, x and y labels, 
             save as an jpg image, 
             #show the image in the console to check if everything is right
    return: None
    '''
    mean_score = list(mean_d.values())
    median_score = list(median_d.values())
    max_score = list(max_d.values())
    min_score = list(min_d.values())
    if list(mean_d.keys()) == list(median_d.keys()) == list(max_d.keys()) == list(min_d.keys()):
        x_labels = list(mean_d.keys())
    else:
        print(f"{word}labels not matching")
        return None
    
    plt.figure(figsize=(15, 10))
    # Plotting the mean line with label
    plt.plot(x_labels, mean_score, label='mean', marker = 'o', linestyle='-')

    # Plotting the median line with label
    plt.plot(x_labels, median_score, label='median', marker = 'o', linestyle='-')

    # Plotting the max scatter with label
#     plt.plot(x_labels, max_score, marker = 'o', linestyle='-', label='max')
    
    # Plotting the min sctatter with label
#     plt.plot(x_labels, min_score, marker = 'o', linestyle='-', label = 'min')

    plt.legend()
    plt.title(f'Sentiment score: {word}')
    plt.xlabel('Years')
    plt.xticks(rotation=75)
    plt.ylabel('Score')
    plt.grid(True)
    fig = plt.gcf()
    fig.savefig(f'visualization/LineChart_{word}.jpg', bbox_inches='tight')
    plt.show()
    
    
```
